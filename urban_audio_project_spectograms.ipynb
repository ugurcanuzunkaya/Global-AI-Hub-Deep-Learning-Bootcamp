```python
# Creating spectograms function
def create_spectrogram(y):
  spec = librosa.feature.melspectrogram(y=y)
  return librosa.amplitude_to_db(spec, ref=np.max)
```

```python
# Opening all fold and class files for photos
for i in range(len(metadata["fold"])):
  path = f"C:/Users/ugurc/Documents/GitHub/Global-AI-Hub-Deep-Learning-Bootcamp/Spectograms_grup2/{metadata['fold'][i]}/{metadata['classID'][i]}"  # before the Github put your path
  with contextlib.suppress(OSError):
    os.makedirs(path)
print("All done")
```

```python
fig, ax = plt.subplots()
for i in range(len(metadata["slice_file_name"])):
  y, sr = librosa.load(f"C:/Users/ugurc/Documents/GitHub/Global-AI-Hub-Deep-Learning-Bootcamp/UrbanSound8K/audio/fold{metadata['fold'][i]}/{metadata['slice_file_name'][i]}")
  spectogram_y = create_spectrogram(y)
  img = librosa.display.specshow(spectogram_y, x_axis='time', y_axis='mel', sr=sr, fmax=8000, ax=ax)
  plt.savefig(f"C:/Users/ugurc/Documents/GitHub/Global-AI-Hub-Deep-Learning-Bootcamp/Spectograms_grup2/{metadata['fold'][i]}/{metadata['classID'][i]}/{i}.png")
  print(f"{i} done")
print("-"*30)
print("all done")
```
