```python
# importing essential libraries
import pandas as pd
import os
import librosa
import numpy as np
from tqdm import tqdm
import matplotlib.pyplot as plt
import librosa.display
```

```python
audio_dataset_path = "/content/drive/MyDrive/urban/UrbanSound8K/UrbanSound8K/audio"
metadata = pd.read_csv("/content/drive/MyDrive/urban/UrbanSound8K/UrbanSound8K/metadata/UrbanSound8K.csv")
metadata.head()  # controlling if everything okay
```

We see that most of the data is balanced

```python
def create_spectrogram(y):
  spec = librosa.feature.melspectrogram(y=y)
  return librosa.amplitude_to_db(spec, ref=np.max)
```

```python
import contextlib
for i in range(len(metadata["fold"])):
  path = f"/content/drive/MyDrive/urban/UrbanSound8K/{metadata['fold'][i]}/{metadata['classID'][i]}"
  with contextlib.suppress(OSError):
    os.makedirs(path)
print("All done")
```

```python
fig, ax = plt.subplots()
for i in range(len(metadata["slice_file_name"])):
  y, sr = librosa.load(f"/content/drive/MyDrive/urban/UrbanSound8K/UrbanSound8K/audio/fold{metadata['fold'][i]}/{metadata['slice_file_name'][i]}")
  spectogram_y = create_spectrogram(y)
  img = librosa.display.specshow(spectogram_y, x_axis='time', y_axis='mel', sr=sr, fmax=8000, ax=ax)
  plt.savefig(f"/content/drive/MyDrive/urban/UrbanSound8K/{metadata['fold'][i]}/{metadata['classID'][i]}/{i}.png")
  print(f"{i} done")
print("-"*30)
print("all done")
```
